```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gender_income=df[['Gender','Income']].groupby(by=['Gender']).mean()
gender_income
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(4,4))
bar_colors = ['tab:red', 'tab:blue', 'tab:orange']
plt.bar(gender_income.index,gender_income.values[0],color=bar_colors)
plt.title("Income Per Gender")
plt.xticks(rotation=60)
plt.show()
```

```python
education=df[['Education Level','Income']].groupby(by=['Education Level']).mean()
education
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(3,3))
bar_colors = ['tab:red', 'tab:blue', 'tab:orange','tab:gray']
plt.bar(education.index,education.values[0],color=bar_colors)
plt.title('Income Per Education Level')
plt.xticks(rotation=60)
plt.show()
```

```python
df.info()
```

```python
marital=df[['Marital Status','Income']].groupby(by=['Marital Status']).mean()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(3,3))
plt.bar(marital.index,marital.values[0],color=["deeppink","lightblue","lightgray","lightgreen"])
plt.title('Income Per of Marital')
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x='Risk Rating', data=df, order=df['Risk Rating'].value_counts().index)
plt.title('Count of Risk Rating')
plt.xlabel('Risk Rating')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.countplot(x='Risk Rating', hue='Gender', data=df)
plt.title('Risk Rating by Gender')
```

```python
plt.figure(figsize=(5, 4))
sns.boxplot(x='Risk Rating', y='Income', data=df)
plt.title('Income by Risk Rating')
plt.xlabel('Risk Rating')
plt.ylabel('Income')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.select_dtypes(['float64','int64']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation of Fnancial Data')
plt.show()
```
